# Harvester features

list alle features from Camera and their values

```python
from harvesters.core import Harvester
```

```python
h = Harvester()

h.add_file('C:\\Program Files\\Allied Vision\\Vimba_6.0\\VimbaGigETL\\Bin\\Win64\\VimbaGigETL.cti')
h.update()
```

```python
if h.device_info_list != []:
    ia = h.create() #ia: ImageAquirer Object
    features = dir(ia.remote_device.node_map)
    last_elem = features.index('WidthMax')
    features = features[:last_elem+1]
    for feature in features:
        
        node = ia.remote_device.node_map.get_node(feature)
        if (str(type(node)) == "<class 'genicam.genapi.ICommand'>") or \
            (str(type(node)) == "<class 'genicam.genapi.ICategory'>"):
            print(feature, '\t \t \t',"commmand or category")
        else:
            try:
                value = node.value
                print(feature,'\t \t \t', value)
            except:
                print(feature)
    ia.destroy()
else: 
    print ("no camera detected")
```

```python
ia.destroy()
h.reset()
```
