Gauß für y-Achsen values

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def gauss_func (x, mu, sig):
    return 1./(np.sqrt(2.*np.pi)*sig)*np.exp(-np.power((x - mu)/sig, 2.)/2.)
```

```python
x_werte = np.linspace(-4, 4, 100, dtype=float)
y_werte = gauss_func(x_werte, 0, 3)
#mu Lage des Maximums
#sig variance (Breite)
y_werte
```

```python
plt.plot(x_werte, y_werte, 'b-')
plt.xticks(np.arange(-4,5,1))
```

3D Gauß

```python
def gauss_func_3D (x, y,  mu_x, sig_x, mu_y, sig_y):
    return np.multiply(gauss_func(x, mu_x, sig_x),gauss_func(y, mu_y, sig_y))
```

Plotten mit Funktion die f(x,y) berechent

```python
x_werte = np.linspace(-4, 4, 100, dtype=float)
y_werte = np.linspace(-4, 4, 100, dtype=float)
def f(x,y):
    return gauss_func_3D(x,y,0,1,0,1)
X, Y = np.meshgrid(x_werte, y_werte)
Z = f(X,Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

Plotten mit allen drei Datensätzen gegeben

```python
from numpy.random import randn
```

```python
N = 1000
x_werte = np.array(randn(N),dtype=np.float32)
y_werte = np.array(randn(N),dtype=np.float32)
x_werte
```
