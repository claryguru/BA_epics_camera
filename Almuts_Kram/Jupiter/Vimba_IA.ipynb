# Vimba Image Aquirer

- choose camera by idea
- in a loop get current frame and save it as a np.array file 

```python
#Import
from vimba import *
import numpy as np
```

```python
#Main
with Vimba.get_instance () as vimba:
    cams = vimba.get_all_cameras ()
    if len(cams) > 0: 
        with cams [0] as cam:
            frame = cam.get_frame ()
            print(cam.)
        np_a = frame.as_numpy_ndarray()
        print(type(np_a))
    else: print("no camera detected")
    
    
```
