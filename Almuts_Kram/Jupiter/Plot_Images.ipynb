```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.ndimage import rotate
```

```python
def open_plot (nr):
    filename = 'D:\\HZB\\Camera_Data\\mls13\\'+ str(nr) + 'test.npy'
    with open(filename, 'rb') as f:
        ar = np.load(f)
    return ar
    
```

```python
ar = open_plot(1)
tr = rotate(ar, 20)
print(type(tr))
plt.imshow(tr, cmap='tab10')
plt.show()
```

```python
#jedes nte image aus Order plotten




for i in range (1,100,3):
    ar = open_plot(i)
    plt.imshow(ar, cmap='tab10')
    plt.show()
```
