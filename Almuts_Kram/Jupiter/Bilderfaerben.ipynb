```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
im = cv2.imread('D:\\HZB\\BA_epics_camera\\Thesis\\images\\pinnhole.png')


print(type(im))
print(im.shape)
print(im.dtype)

plt.imshow(im, cmap='viridis')
plt.show()
```

```python
def rgb2gray(Img):
    row,col,ch = Img.shape
    g = [ ]  
    for i in range(row) :
        a = []
        for j in range(col):
            a.append(Img[i,j,0])
        g.append(a)
    gr = np.array(g)
    return gr

grIm = rgb2gray(im)
print(grIm.shape)
print(grIm)
plt.imshow(grIm, cmap='viridis')
plt.show()
```
