```python
from vimba import *
import numpy as np
from matplotlib import pyplot as plt
import time
```

```python
with Vimba.get_instance () as vimba:
    cams = vimba.get_all_cameras ()
    with cams [0] as cam:
        # Aquire single frame synchronously
        while True: 
            frame = cam.get_frame ()
            np_a = frame.as_numpy_ndarray()
            plt.imshow(np_a, cmap='tab10')
            plt.show()
            time.sleep(2)
```
